### 5_2. 중복 데이터 처리

```python
df_dup = pd.DataFrame({
    'c1' : ['a','b','c','a','b'],
    'c2' : [1,2,3,1,2,]
})

df_dup  # row가 중복된걸 중복됬다고 함~~~
```

```python
dup_rows = df_dup.duplicated() #boolean 반환 -> 즉, indexing 가능
# dup_rows(dup_rows) => 에러남
```

```python
df_dedup = df_dup.drop_duplicates(keep="fist") # 중복된 행 제거
df_dedup = df_dup.drop_duplicates(keep="last") # 인덱스가 바뀜
#근데 제거는 조심해야함~!

df_dedup
```

### 통계기반 데이터 종류 : 정성적, 정량적 Data
- 정성적:Qualitative , 질 O ,양 X => 숫자아닐수도 있음, 문자도 Ok
  
  1.**명목형 Data** : 순서나 크기가 의미X, 단순히 이름붙이거나 분류에 사용 

      ex) 성별,혈액형
    
  2.**순서형 Data** : 순서가 의미 O , 간격은 의미X

      ex)학점A,B,C 만족도(매우,불,만,) => 간격값은 알 수 없다.
- 정량적:Quantitative , 양 O ,질 X => 숫자
  
  1.**이산형(discrete)** : 셀 수 있는 정수 형태 
  
       ex)주사위

  2.**연속형**
   :연속,소수점 값
      
      ex) 키, 몸무게
