```python
# pd.cut() : 연속형데이터 -> 범주형 데이터로 변환
# df_age['컬럼명'] : 없으면 컬럼명 설정
```

```python
import pandas as pd
import numpy as np
```

```python
df_age = pd.DataFrame({
   'age':[5,15,21,28,39,42,58,65] 
})

df_age
```

```python
bins = [0,19,29,39,49,69]
labels = ['10대이하','20대','30대','40대','50대이상']
```

```python
df_age['연령대'] = pd.cut(
    df_age['age'], #나이데이터 기준으로 연령대 나누겠다
    bins=bins, #구간경계값 리스트
    labels=labels,  # 각구간에 붙일 이름(라벨) 지정
    right=True # 오른쪽 경계 포함함
    # right=False -> 39살이 40대로 처리됨
)

df_age
```

### 6.데이터프레임의 다양한 응용

```python
#그룹연산
# 3.1 그룹객체 만들기(분할 단계)
df_group = pd.DataFrame({
    '팀' : ['개발','마케팅','개발','영업','마케팅','개발'],  
    '이름' : ['민준','서연','도윤','예은','시우','지우'],
    '실적' : [120,90,150,110,130,100]
})
df_group
```

```python
grouped = df_group.groupby("팀")
print(grouped)
for name, group in grouped:
    print(name)
    print(group) # 데이터 프레임이 나옴
    
```

```python
#각 팀의 실적 합계 구하기
sum_by_team = grouped['실적'].sum()
print(sum_by_team)
mean_by_team = grouped['실적'].mean()
print(mean_by_team)
max_by_team = grouped['실적'].max()
print(max_by_team)
```

```python
agg_result = grouped['실적'].agg(['sum','mean','max','count']) #DB에서 이거할려면 여러번 계산해야함
agg_result
```

```python
# 데이터프레임 합치기 : concat
# 병합merge, join과 비슷
```

```python
df1 = pd.DataFrame({
    'A':['A0','A1'],
    'B':['B0','B1']
})

df2 = pd.DataFrame({
    'A' : ['A2','A3'],
    'B' : ['B2','B3'],
    # 'C' : ['B2','B3'], 만약 B가 아니라 C로 쓰면 새로운 행 생김, outer join, A,B,C 다나오니까 outer join

})
df2
```

```python
# 위아래로 연결(행방향, axis = 0)
result_row = pd.concat(
    [df1,df2],
    axis = 0, #default
    ignore_index = True, # 기존인덱스 무시해라 / default = False(즉, 기본인덱스 써라)     
    # join='inner' # 중북된 A만 나옴/ 앞으로 아주 필요함~
    # join='outer' # 디폴트로 처리, 다 나옴~ / join=outer없으면 그냥 다나옴~~~ 그게 그냥 디폴트니까
)
result_row
```

```python
df3 = pd.DataFrame({
    'C' : ['C0','C1'],
    'D' : ['D0','D1']
},index=[1,2]) 
df3
pd.concat([df1,df3], axis=1) # 옆으로 붙일땐 인덱스의 영향을 받는다.
# concat: 리스트로 들어감(즉, 여러개 들어갈수 있음~~~)
#merge: 데이터 값가지고 붙임
#concat: 기존의 컬럼의 값 과 인덱스의 값 붙임

print(df1)
print(df2)
print(df3)
print(pd.concat([df1,df3], axis=1))
```

```python

# merge 랑 concat 비교해야함
# merge :
# 1. 고객정보
df_cust = pd.DataFrame({
    "고객아이디":[1,2,3,4],
    '이름' : ['민준','서연','도윤','예은']
})
df_cust
```

```python
# 구매정보
df_purcha = pd.DataFrame({
    '고객아이디' : [2,3,3,5],
    '구매상품' : ['노트북','모니터','키보드','마우스']
})
df_purcha
#여기서 concat 하면 매치가 안맞듬
#고객아이디랑 컬럼의 값으로 붙여야함 :merge

#조인조건: 고객아이디
#데이터 프레임 2개만 들어갈수 있음~~~!!!!
#default = inner , 조건에 맞는건만 가져옴~
#양쪽 데이터 프레임 컬럼명 같을 경우 : ON
#컬럼명 틀릴때 -> left,right 컬럼명 따로따로 지정
df_merged = pd.merge(
    df_cust, df_purcha, # 왼쪽:cust, 오른쪽 :구매상품
    left_on = '고객아이디',
    right_on = 'custid',
    # on = '고객아이디', # 양쪽 데이터 프래임이 다 고객아이디니까 on사용가능~~~
    # how = 'inner' # inner default , 같은것만 나옴
    how = 'outer' # outer -> full outer
    # how = 'left' -> left outer: 고객 다나옴
    # how = 'right' # right outer : 구매 다나옴
    , suffixes=('_cust','_purcha')
)
df_merged
```

```python
# 피벗테이블 만들기
df_long  = pd.DataFrame({
    '월': ['1월','1월','2월','2월','1월','3월'],
    '제품':['A','B','A','B','A','B'],
    '판매량': [100,150,120,180,50,70]    
})
#피벗 : 월별, 상품별 집계
df_long
```

```python
df_pivot  = df_long.pivot_table(
    index='월' #인덱스로 보낼 컬럼 지정(그룹)
    , columns = '제품' #컬럼으로 보낼 컬럼 지정(그룹), 제품별
    , values = '판매량'# 집계에 사용할 데이터 지정
    , aggfunc='count' #'mean' default / <주의> 갯수없으면 0개 아니라 nanX
    , fill_value=0
)
df_pivot #각 결과는 평균값
```
