```python
# iris data loading
import pandas as pd
import numpy as np

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
```

```python
# 1. 데이터로딩
iris = load_iris()
X = iris.data  # 컬럼명이 없다.
y= iris.target # 숫자 [0,1,2]

# 컬럼명, 품종명 -> 나중에 시각화에 넣기 위해서
feature_names = iris.feature_names
class_names = iris.target_names
print(feature_names)
print(class_names) # class name 이 빠지면 결정트리가 무엇으로 되어있는지 알 수 없음
```

```python
tree_model = DecisionTreeClassifier(random_state=42)

# 학습
tree_model.fit(X,y) # 시각화가 먼저여서 이것저것 다른 코드 뺀거임
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
plt.figure(figsize= (20,10))

plot_tree(
    tree_model,
    feature_names = feature_names,
    class_names=class_names,
    filled = True,
    fontsize = 10   
)
 
plt.savefig('decision_tree.png') # 파일따로 저장가능함 
plt.show()
```
