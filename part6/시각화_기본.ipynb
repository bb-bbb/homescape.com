```python
#1.Axes 가 없는 빈 Figure 생성 -> 축 안생김
fig = plt.figure()

#2.단일 Axes를 포함하는 Figure 생성
fig, ax = plt.subplots()

#3.2X2 그리드 형태의 Axes를 포함하는 Figure 생성
fig, ax = plt.subplots(2,2) # 2X2 -> 축이 4개 나온다.
plt.show() 

#4.복잡한 레이아웃의 Axes를 포함하는 Figure 생성
fig, ax = plt.subplot_mosaic ([
    ['left','right_top'],
    ['left','right_bottom']
])
plt.show()
```

```python
# 위 데이터를 시각화(산점도: 데이터 분포 확인차 사용)
# X = 'a', y = 'b',  color = 'c' , (점크기)size = 'd'
fig,ax = plt.subplots(
    figsize  = [5,2.7]
    )
ax.scatter(x='a'
        ,  y='b'
        ,  c='c'
        ,  s='d'
        , data=data)

plt.show()

```

```python
#코딩스타일 2개존재
#1.객체지향 스타일(OO)

x = np.linspace(0,2,100) # 0~2까지, 100개분할해서 만들어라
fig,ax = plt.subplots(figsize=(5,5))
ax.plot(x,x) #x,x같으면 직선~
ax.plot(x,x**2)
plt.show()
ax.plot(x,x**3)
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('Simple plot')
plt.show()
#2차함수일때: <주의> x값이 0~1사이올때 제곱하면 값이줄어듬

```

```python
#plot style, 위에 코드랑 결과는 같지만 위에는 조립해서 쓴거임
plt.figure(figsize=(5,2.7)) #축 없는 빈 도화지 생성

plt.plot(x)
plt.plot(x**2)
plt.plot(x**3)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple plot")
plt.show()
```

```python
##헬퍼 함수 만들기 : 재사용 위해!!!
##재사용위해서 "객체코딩스타일 적용한거인~"
```

```python
def my_plotter(ax, data1, data2, param_dict):
    out = ax.plot(data1, data2, **param_dict)
    return out
```

```python
data1, data2, data3, datat4 = np.random.randn(4,100) # 아래 4줄짜리를 한줄로 끝내버림,pandas에만 있으
# np.random.randn(2,3) # 2행 3열
# print(len(data1))

# data = np.random.randn(4,100)

# data1 = data[0]
# data2 = data[1]
# data3 = data[2]
# data4 = data[3]
```

```python
data1,data2,data3,data4 = np.random.randn(4,100)
# 1X2 서브플롯
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(5,2.7)) # 1,1이 디폴트
my_plotter(ax1,data1,data2,{'marker':'X'})
my_plotter(ax2,data3,data4,{'marker':'o'})
# plt.subplots(1,2,figsize=(5,2.7)) # 그래프2개나옴
plt.show()
```
