```python
# 기본 설정 및 브라우저 실행 
from selenium import webdriver # 브라우저별 제어 방식 , 크롬 드라이버, 엣지드라이버, safari
from selenium.webdriver.chrome.service import Service as ChromeService # 크롬 브라우저 제어
from selenium.webdriver.chrome.options import Options 

# 선택
options = Options()

# chrome 드라이버 서비스 실행하도록
service = ChromeService()
driver = webdriver.Chrome(service=service , options=options)

driver.get('https://www.naver.com')
#링크열면 : Chrome이 자동화된 테스트 소프트웨어에 의해 제어되고 있습니다!
#브라우저에 모든기능 제어 가능~~ with 코드로
```

```python
driver.close() #브라우저 닫힘
```

#### selenium 정리
1. 브라우저 제어
    - 탐색\
        -get(url:str) : 지정된 url로 이동,페이지가 완전히 로드될때 까지 기다려야함(wait)\
        -back(): 뒤로 가기버튼 클릭\
        -forward(): 앞으로\
        -refresh(): 새로고침\
        -current_url() : 현재 페이지의 url 주소 

2. 웹 요소 탐색
    - 단일 요소 찾기\
        -**driver.find_element(by:By , value:str)**: 첫번째 한개 반환\
            -By.ID, By.CLASS_NAME,  By.TAB_NAME, BY.CSS_SELECTOR, By.Xpath\
            -xpath: f15 코드에서 정확한 위치 파악
    - 여러 요소 찾기\
        -driver.find_elements(by:By, value:str):리스트반환


3. 웹 요소 조작 및 정보 추출 : 웹 요소 객체를 선택하고 조작한다.
    - element.click()  ex)코드클릭할 수 있다.
    - element.send_keys(*value) : 선택한 요소(text,select) / value가 들어간곳은 다 됨~
    - element.clear(): value삭제
    - element.submit: submit제출
    - element.text(속성): 화면에 표시된 텍스트 반환
    - element.get_attribute(name) : 태그에 있는 속성의 값을 반환

    


```python

from selenium.webdriver.common.by import By

#네이버에서 로그인 화면에서 ->copy -> xpath
search_id = driver.find_element(By.XPATH , '//*[@id="id"]')
search_id.send_keys('내아이디') # 내아이디라는 문자열을 로그인 칸에 넣어라
```

```python
#비밀번호 넣기
search_pw = driver.find_element(By.XPATH,'//*[@id="pw"]')
search_pw.send_keys('1234')
```

```python
# 로그인넣기

search_login = driver.find_element(By.XPATH,'//*[@id="log.login.text"]')
search_login.click()
```
