```python
import requests               # 모듈
from bs4 import BeautifulSoup # 클래스 



# https://www.chicagomag.com/Chicago-Magazine/November-2012/Best-Sandwiches-Chicago/


url_base = 'https://www.chicagomag.com' #com 이후에 하나라도 더 들어가면 오류다.
url_sub = '/Chicago-Magazine/November-2012/Best-Sandwiches-Chicago/' # 메인페이지 
url_main = url_base + url_sub
# url_main

headers = {
    'User-Agent' : 'Mozilla/5.0'
    
}

# header 설정하는 이유
# headers: 웹 서버에 요청을 보낼 때 함께 전달되는 정보
# 'User-Agent' :요청을 보내는 프로그램의 정체
# 'Mozilla/5.0' : 일반적인 브라우저 식별자


#리퀘스트 날리면 된다.
response = requests.get(url=url_main, headers=headers)
# -->웹 서버가 이 요청을 브라우저에서 온 것처럼 인식해서 데이터를 더 잘 반환
# requests.get : HTTP GET 요청을 보내는 함수,웹 페이지나 API에서 데이터를 가져올 때 사용
# print(response.text[:1000])
# headers없을때  결과: 403오류

# ============main page text 가져오기===================

# 2. 파싱 = DOM tree 생성
# soup = BeautifulSoup(response,'html.parser')
# 에러 : BeautifulSoup(response,'html.parser') reponse 에 텍스트 넣어야함 object 넣으면 X

soup = BeautifulSoup(response.text,'html.parser') #text: html code, parsing

```

```python
# 연습용 : find로 하나태그 가져오기

# div태그에 sammy 클래스 가져와라, 근데 그냥 find니 하나만 가져와라
# <주의> class_ : 언더바사용해야함
soup.find('div',class_="sammy") # 코드에서 클래스 더블클릭

# 결과
# <div class="sammy" style="position: relative;">
# <div class="sammyRank">1</div>
# <div class="sammyListing"><a href="/Chicago-Magazine/November-2012/Best-Sandwiches-in-Chicago-Old-Oak-Tap-BLT/"><b>BLT</b><br>
# Old Oak Tap<br/>
# <em>Read more</em> </br></a></div>
# </div>

tmp = soup.find('div',class_="sammy")
print(tmp)

# 2.rank 뽑기
print(tmp.find(class_='sammyRank').text)
# tmp.find(class_='sammyRank') : <div class="sammyRank">1</div> 
# .text 위에 <div class="sammyRank">1</div> 에서 텍스트만 뽑아라

# 3. href에서 value 값 가져오기 with class 선택자로
print(tmp.find(class_='sammyListing').text)
# 결과 :'BLT\nOld Oak Tap\nRead more ' -> 리스트로 구분한후 0번째량 1번째 가져오면 됨~
print(tmp.find(class_='sammyListing').text.split('\n'))
# 결과:['BLT', 'Old Oak Tap', 'Read more ']
print(tmp.find(class_='sammyListing').text.split('\n')[0:2])


# 4.url 뽑기 즉,href뽑아라
# a태그 찾아라
print(tmp.find('a').get('href')) # get method 방식이 더 안정적
print(tmp.find('a')['href'])

# 전체 뽑기
print(url_base + tmp.find('a').get('href'))
```

```python
# 메인에서 50개 접근 데이터 추출
import requests
from bs4 import BeautifulSoup

headers = {
    'User-Agent':'Mozilla/5.0'
    
}

response = requests.get(url=url_main, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# 메인에서 데이터 추출 : 순위, 메인메뉴, 카페명 ,상세URL -> 각각 50개씩 저장

ranks= [] # 순위
main_menus=[] # 메인메뉴
cafe_names = [] # 카페명
url_adds = [] # 상세화명 URL


list_sammy = soup.find_all('div',class_='sammy')
# print(len(list_sammy))

for item in list_sammy: # item == tmp ,하나씩 꺼내니깐 itemp 은 temp 다
    ranks.append(item.find(class_='sammyRank').text) # 순위 추출해서 리스트 추가
    tmp_str = item.find(class_ = 'sammyListing').text.split('\n')
    main_menus.append(tmp_str[0]) #메인메뉴
    cafe_names.append(tmp_str[1]) #카페명 추가
    
    
    tmp_url = item.find('a')['href'] # https:// 시작 또는 /Chi 시작도 있음
    
    if tmp_url.startswith('https://'):
        url_adds.append(tmp_url)
    else:
        url_adds.append(url_base + tmp_url)
    
    
    
    
    
# 주의할점: 어떤 샌드위치에는 url에http 가 있고 어떤거는 없다. 함부러 가져가지 말라고 소스 수정할 수 있음
# 웹페이지가 일관적이지 않다~~~를 알아야 함
```

```python

#메인메뉴에서 필요한 데이터 다 수집함~~~너네가 코드 꼬아도 괜찮아~~~
print(ranks[:5])
print(main_menus[:5])
print(cafe_names[:5])
print(url_adds[:5])
url_adds[:25]
```

```python
import pandas as pd
import numpy as np

data = pd.DataFrame({
    'Rank':ranks,
    'Menu':main_menus,
    'Cafe':cafe_names,
    'Url':url_adds
})

data.head(10)

# https://github.com/github/spec-kit : 분석설계 물어보변서 코드짤수 있음
# 위에 speckit ->cursor, copilot 적용
# speckit: 프로젝트 사이즈 중이상될때
# 분석설꼐없이 AI 사용하면 X
# 보안문제 해결 ->Docker사용
# <전체적 프로젝트 흐름>
# 데이터분석엔 분석설계가 필수 -> 엑셀사용 -> 팀원들한테 나눠서 -> 수집한데이터를 DB에 넣도록
```

```python
# 1등 페이지에서 가격, 전체 주소 추출
data['Url'] # URL 이 시리즈로 나오고, 인덱스로 접근가능 / 시리즈 : 반복가능
url_str = data['Url'][0]
url_str 
```

```python
headers = {
    'User-Agent' : 'Mozilla/5.0'   
}
response = requests.get(url=url_str, headers = headers)

# 2.헤더 가져온후 파싱하기
soup = BeautifulSoup(response.text , 'html.parser')
tmp_str = soup.find('p',class_='addy').text # 1번샌드위치  에서 가격있는 tag 소스가져오기
print(tmp_str)

```

```python
# 공백을 기준으로 나누기
tmp_str.split()

# 나뉜거에서 가격 0 만 가져오기
tmp_str.split()[0][:-1] # 10.에서 .빼기   #가격추출

# 주소만 뽑기
tmp_str.split()[1:-2] # join
tmp_str.split()[1:-2] # 공백포함해서 join하기  #주소추출
' '.join(tmp_str.split()[1:-2]) # 공백포함해서 join하기 -> 리스트를 하나문자열로 만들기




```

```python
# tqdm 왜사용 
# 데이터 전처리를 할 때 상태확인을 위해 자주 사용, 언제끝날지 알기때문에 커피라도 먹을 수 있음

import time 
from tqdm import tqdm # tqdm 모듈안에서 tqdm들어감

for number in tqdm(range(10)):
    time.sleep(1)
    print(number)
    
print('데이터 수집 완료')
```

```python
# 상세페이지 50개 반복처리
# 가격, 주소 저장
price = []
address = []

# data에서 인덱스0~50까지 가져오기
# for idx in tqdm(data.index):
# data.index


for idx in tqdm(data.index):
    url_str = data['Url'][idx]
    
    headers = {
        'User-Agent' : 'Mozilla/5.0'
    }
    response = requests.get(url=url_str, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    get_str = soup.find('p',class_='addy').text
    price.append(get_str.split()[0][:-1])
    address.append(' '.join(get_str.split()[1:-2]))
    
# 근데 pqdm 사용하면 속도 느려짐
```
