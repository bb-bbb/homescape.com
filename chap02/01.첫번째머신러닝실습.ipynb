    - 모델에서 예측한값이 -  원래값 = 오차
    - 오차로 모델을 수정함


- [3]<예측>
    - 모델을 업데이트 함


    - 여기까지 : model (데이터준비~예측까지 계속반복함, for 오차가 최소될때까지)
    - ((반복))종료되면 : 모델완성

    - 1.학습에 사용되지 않은 new data

    - 3. 위에 결과로 평가함 with 예측(label값)

- [4]<평가>
    - <추가> 편차, 오차를 가지고 모델이 얼마나 성능이 좋은지 판단
    - 잘못나오면 [1]데이터준비부터 다시해야함

    - seed : random_state = 11 : 콩심은데 콩나온다~(헷갈릴수 있음)
    - 여기서 11은 의미는 없다 그저 랜덤값/ 안바꾸면 됨 10,200하던 값을 바꾸지만 않으면 된다.
