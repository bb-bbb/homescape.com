```python
df_scores = pd.DataFrame({
    '국어':[85,92,78,95,88],
    '수학':[90,88,91,76,82],
    '영어':[88,95,80,92,85],
    
}, index = ['민준','서연','도연','예은','시우'])
df_scores
```

```python
# 과목별 평균(열 기준)
avg_by_subject = df_scores.mean() # 시리즈로 결과가나옴, 즉 인덱스뽑을 수 있다
avg_by_subject['국어']
avg_by_subject


#describe으로도 접근가능 ,근데 2번 처리해야함
df_desc =df_scores.describe()
df_desc.loc['mean']['국어']
```

```python
#학생별 평균점수(행기준, axis=)
#<중요>axis=0 , default 아래방향으로 평균구해라~
#<중요>axis=1  옆방향으로 평균구해라
avg_by_students=df_scores.mean(axis=1)
avg_by_students
```

```python
#중앙값,최대,최소
mid_by_students=df_scores.median() # 할수는 있지만 그리 의미없는 작업
mid_by_students
```

```python
#최대,최소
df_scores.max(axis=0) # 열기준:과목별, 해당 열의 최대값
df_scores.max(axis=1) # 행기준:학생별, 학생들에서 최대값
df_scores.min()#axis=0

#상관관계를 구하는게 : 공(두개,co)분산
#분산(평균으로부터 얼마나 퍼져 있는지): 한개변수, 공분산: 두개 변수
#수치X, 방향O
#양수: 두변수가 양의방향 관계 = 두변수 완전히 똑같다
#0: 관계없다
#음수: 음의방향 관계 = 완전히 반대

#<결론>기울기과 관계다!!!딥러닝은 기울기 찾는거다!!

#0.75넘어가면 관계크다
```
