```python
#데이터프레임에서 인덱싱,슬라이싱,팬시인덱싱,블린인덱싱 =>4가지~어마무시하게 중요함
#판다스에서 가장기본 & 가장 중요한

student_data = { 
    '이름' : ['민준','서연','도윤','예은','시우'],
    '나이' : [18,17,18,17,18],
    '도시' : ['서울','부산','서울','인천','부산'],
    '국어' : [85,92,78,95,88],
             }

#student_data['이름']
df = pd.DataFrame(student_data, index=['A','B','C','D','E'])
df

```

```python
# 여러개 컬럼 선택 : 슬라이싱
# 1.슬라이싱 : 연속된 여러개 => LOC or iLOC(행먼저 접근, 열 접근) 사용 , 대괄호 X
# print(df['이름':'나이']) =>콜론 사용 X , :는 범위 의미


# 2.팬시인덱싱 : 비연속된 여러개 , 대괄호 여러개
#==>리스트: 컬럼 여러개 지정 가능
df[['이름','국어']] # 내가 원하는 컬럼만 딱 선택할때 진짜 많이 사용함~
df[['국어','이름']]


```

```python
# 슬라이싱 : 특정 범위의 데이터를 잘라내어 선택하는 방법, 행과 열을 자를 수 있다.
# 행먼저 접근, 열 접근

#행인덱싱 : loc(문자[레이블]기반), iloc(integer location,숫자 기반), 보통은 loc사용
#기본구조: loc[행,열] => 행선택하고 열 선택

# 	이름	나이	도시	국어
# A	민준	18	서울	85
# B	서연	17	부산	92
# C	도윤	18	서울	78
# D	예은	17	인천	95
# E	시우	18	부산	88

#위테이블에서 B~D만 가져오고 싶음

print(df.columns)
df.loc['B':'D'] # 팬시 인덱싱






```
