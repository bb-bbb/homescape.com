```python
# 데이터프레임에서 인덱싱, 슬라이싱, 팬시인덱싱, 블린인덱싱
student_data = {
    '이름' : ['민준','서연','도윤','예은','시우'],
    '나이' : [18,17,18,17,18],
    '도시' : ['서울','부산','서울','인천','부산'],
    '국어' : [85,92,78,95,88]
}
# student_data['이름']
df = pd.DataFrame(student_data, index=['A','B','C','D','E'])
df
```

```python
# 1. 기본 인덱싱 : 특정 열이나 행 하나를 선택하는 방법, [] 연산자 사용
# 1.1 컬럼 인덱싱 
print(df['이름'])
```

```python
# 여러개 컬럼 선택
# 1. 슬라이싱 : 연속된 여러개 => loc, iloc을 사용해야 한다.
# df['이름':'도시'] 이건 안된다

# 2. 팬시인덱싱 : 비연속된 여러개 
df[['국어','이름']]
```

```python
df.loc['B':'D']
df.loc[['B','C','D']] # 팬시인덱싱, 팬시인덱싱은 대괄호
df.loc['B':'D',['이름','나이','도시']]
df.loc['B':'D','이름':'도시']  # 슬라이싱 해서 머신러닝,딥러닝 집어넣어야함, 콜론(:) = 연속된 범위
df.loc[:,:] # 전체슬라이싱(:연속 범위 지정),행(start:end), 열(start:end) 선택
#리스트랑 다른점: 마지막 end-1을 하지 않는다.레이블 인덱싱 처리를 해서 그럼

#복습: 리스트에서 콜론 들어가는 4가지
#start:end, X:end,start:X, X:X

df.loc['C':,:'도시'] # 도윤부터 다 가져와라~
#파이썬 2장이 중요함~~

df.loc[['B','D'],['이름','국어']] # 팬시색인, 
#주로: 컬럼은 비연속, 로우는 연속처리
df.loc[:,['이름','국어']] #전체 가져올꺼면 대괄호 없애야함 ,이런경우가 많음
```

```python
# 정리
# 1.LOC[행,열] : 레이블 기반으로 행과 열 선택
# 1.iloc[행,열] : 정수인덱스 기반으로 행,열 선택
# 3.query (문자열 조건식)
query = '도시=="서울" and 국어 >= 80' #컬럼은 변수명 , 문자열 처리 X , 바깥에 작은따옴표, 안에는 큰따음표 !형식 지키기
df.query(query)

# 4. isin(리스트) # 너무중요!! 리스트로 쌓아야 함!!!!, 한개일지라도 list 로 감싸기!!!!!!!!!! =>return boolean Series
# 서울,인천 학생 조회
# df['도시'].isin[('서울','인천')]
df['도시'].isin(['서울','인천']) # 불리언으로 먼저 뽑고 그걸 아래 df에다가 놓는다.
df[df['도시'].isin(['서울','인천'])]

#5.filter : 특정 컬럼 선택
df.filter(like='시') # regex:정규표현식을 줄 수 있다, 컬럼명 검색
df.filter(like='이') # regex:정규표현식을 줄 수 있다, 컬럼명 검색






```
