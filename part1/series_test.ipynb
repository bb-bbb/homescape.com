```python
#Series attribute, method 정리
import pandas as pd
import numpy as np

#리스트로 생성
s = pd.Series([1,2,3,np.nan,6,8], name ='numbers')
#name ='numbers' : 데이터프레임에서 컬럼명
#np.nan : Not a Number을 의미하는 실수형(float) 값
```

```python

#Series object 주요 속성 정리
# 속성: 특성이나 정보를 저장하는 역할
# 1.values : 데이터 값을 ndarray 반환
# 2.index : 인덱스 반환 (RangeIndex, Index)
print(s.index)
print(s_s_n.index)

# RangeIndex(start=0, stop=6, step=1) : 숫자 인덱스
# Index(['민준', '서연', '도윤', '예은', '시우'], dtype='object'):문자인덱스
# 3.dtype: 데이터 타입 반환

print(s.dtype)
print(s_s_n.dtype)

#4.name : 이름을 반환,데이터프레임에서는 컬럼명으로 사용된다
print(s.name)

#5.size : 데이터의 총 갯수 반환 (NaN도 포함, DB와는 다름)
print(s.size)

#6.shape : 형태(shape)를 반환(튜풀), 1차원:rotn
print(s.shape)
#7.ndim:차원수 반환(1반환)
print(s.ndim)
#8.hasnans : 결측치가 하나라도 있으면 True, 없으면 false
print(s.hasnans)
```

```python
# 고객 등급 시리즈
grades = pd.Series(['Gold','Silver','Bronze','Gold','Diamon'])
# grades
# 등급을 점수로 변환하기 위한 딕셔너리 선언
grade_to_score = { 
  'Gold':30,
  'Silver':20,
   'Bronze':10,
                  
}
print(grades)
print(grades.map(grade_to_score)) # 딕셔너리에 해당값없으면 NaN
#map : 데이터를 내가 원하는값으로 바꿀때 엄청많이 쓰임
#다시한번기억 :method 는 object안에 있음
print(grades.map(grade_to_score).fillna(grades)) #fillna:채워라
#indext 역할: 같은 값을 채워라~~ , 키가지고 inner join
```

```python
#4.데이터 정렬 및 변경
# 4.1 sort_values():값을 기준으로 데이터 정렬(ascending=False로 내림차순 가능)
print(s.sort_values(ascending=False)) # NaN이 맨밑에 깔림
print(s) #<주의> 원본은 안바뀐다!!!!!, View다, inplace=False

#view: 데이터를 복사하지 않고 원본 배열의 "다른 관점"을 제공하는 객체
print(s.sort_values(ascending=False, inplace=True)) #inplace=True:원본정렬, 원본값이 바뀌니까 none 출력됨,
print(s) # 원본이 바뀜

#4.2 sort_index(): 잘안씀,인덱스를 기준으로 정렬
#4.3 reset_index(): 많이 사용, 인덱스 초기화, 0부터 새로 인덱스 생성 처리, 데이터 합쳤을때 많이 사용함
```
