```python
#데이터 프레임에서 인덱싱,슬라이싱,펜시 인덱싱, 블린인덱싱

student_data = { 
    '이름' : ['민준','서연','도윤','예은','시우'],
    '나이' : [18,17,18,17,18],
    '도시' : ['서울','부산','서울','인천','부산'],
    '국어' : [85,92,78,95,88],
             }

student_data['이름']
print(type(student_data))
print(type(student_data['이름']))

df = pd.DataFrame(student_data, index=['A','B','C','D','E'])
df

#1.기본인덱싱: 특정 컬럼(열)을 선택하거나, 행하나를 선택하는 방법, []연산자 사용
#1.1 컬럼 인덱싱, 근데 인덱스(A~E)같이 나옴
print(df['이름'])
```
