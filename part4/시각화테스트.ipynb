```python
import matplotlib.pyplot as plt
```

```python
# 데이터 준비
months = ['1월','2월','3월','4월','5월','6월']
visitors = [1200,1350,1500,1420,1680,1800]
```

```python
#시각화
#<준비> 폰트먼저 만들어야 한글 안깨짐, 단 도화지 만들기전(pit.figure)에 해야함!!!!!
plt.rcParams['font.family']='Malgun Gothic'
plt.rcParams['axes.unicode_minus']= False
#도화지 준비
plt.figure(figsize=(10,6))
plt.show() # show: 화면에 표시 , 하얀색 도화지준비

# 선그래프
plt.plot(
    months, #x축 데이터 지정
    visitors, #y축 데이터 지정
    marker='s', #점의 모양 지정,'o':원 , 's':사각형
    linestyle = '--', # 점선으로 바뀜 , dash1 -:실선, dash2 --:점선dashed, : 진짜점선dotted
    color = 'k',
    label="방문자수" #범례표시용

    )

#그래프 제목!
plt.title("월별 웹사이트 방문자수",fontsize=16) 

#x축 그래프 아래 표시
plt.xlabel('월')

#y축 그래프 표시
plt.ylabel('방문자 수')

#정확한 수치 표시
plt.grid(True, linestyle = ':')

# 범례표시
plt.legend()

plt.show()
```

### 막대그래프: 카테고리 데이터(범주형 데이터)의 크기 비교

```python
# plt.bar() : 수직막대 
# pit.barh():수평막대

# 데이터준비
departments = ['인사팀','개발팀','마켕팅팀','디자인팀','영업팀']
employee_counts = [5,15,8,6,12]

```

```python
#도화지
plt.figure(figsize=(5,3))
# 막대그래프
colors = ['lightyellow','lightgreen','lightblue','blue','blue']
plt.bar(
    departments,
    employee_counts,
    color=colors
)

# 그래프 꾸미기
plt.title('부서별 사원 수')
plt.xlabel('부서')
plt.ylabel('사원수')
plt.grid(True, linestyle=':')
plt.ylim(0,18) # y축 범위 설정
```
