```python
# 1. 파이썬 코드에서 Matplotlib 폰트 설정!!
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
from datetime import timedelta
from itertools import combinations
from collections import Counter

# 폰트 설정
plt.rc('font', family='Malgun Gothic')
# 마이너스 부호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

print("한글 폰트 설정이 완료되었습니다.")

# 3. 데이터 로드 및 통합 (모든 문제 풀이의 시작점)
try:
    orders_df = pd.read_csv('../data/orders.csv')
    payments_df = pd.read_csv('../data/payments.csv')
    products_df = pd.read_csv('../data/products.csv')
    shipping_df = pd.read_csv('../data/shipping.csv')
    customers_df = pd.read_csv('../data/customers.csv')

    # 모든 데이터프레임 병합
    df = pd.merge(orders_df, payments_df, on='order_id', how='left')
    df = pd.merge(df, products_df, on='product_id', how='left')
    df = pd.merge(df, customers_df, on='customer_id', how='left')
    df = pd.merge(df, shipping_df, on='order_id', how='left')

    # 데이터 전처리
    date_cols = ['order_date', 'payment_date', 'join_date', 'shipping_start_date', 'shipping_end_date']
    for col in date_cols:
        df[col] = pd.to_datetime(df[col], errors='coerce')
    df['total_sales'] = df['quantity'] * df['price']
    
    print("데이터 로드 및 통합이 완료되었습니다.")

except FileNotFoundError as e:
    print(f"파일을 찾을 수 없습니다: {e}")
```

```python
# 출력 결과를 보고 코딩하세요
# 1. 주문 날짜 기준으로 '년-월' 컬럼 생성
df['year_month'] = df['order_date'].dt.to_period('M')
#주문날짜 order_date 에서 연도-월(YYYY-MM)형식으로 추출하여 year_month라느 새로운 칼럼 만듬
#.dt:datetime 타입의 컬럼에서 날짜 관련 속성이나 메소드 사용하는 접근자
#.to_period('M'): 날짜를 '월'단위로 변환, ex) ,2023-10-20 -> 2023-10 으로 바뀜


# 2. 월별, 카테고리별 총 매출 집계
monthly_sales = df.groupby(['category', 'year_month'])['total_sales'].sum().reset_index()
# 제품 카데고리와 주문별 그룹화하여 총 매출total_sales 합산->결과를 데이터 프레임으로 만들기
# df.groupby(['category', 'year_month']) : 데이터프레임을 카테고리 와 주문별 기준으로 그룹화함
# ['total_sales'].sum() : 각그룹별로 total_sales값 합산
# .reset_index() : 그룹화 결과는 인덱스가 계층적으로 되어있는데,이를 일반적인 테이블 형태로 인덱스를 초기화해서 다시 데이터 프레임 만들기


# 3. 월별 성장률 계산
monthly_sales['prev_sales'] = monthly_sales.groupby('category')['total_sales'].shift(1)
#데이터프레임을 카테고리기준으로 그룹화하고(화장품, 가구, 식품,)기준으로 그룹화
#['total_sales'] : 각 그룹에서 총 매출 컬럼만 선택
# .shift(1) : 선택한 매출 데이터를 한행씩 아래로 이동 즉, 이전달의 매출을 현재행에 대응시킴
# monthly_sales['prev_sales'] :위에 계산된 이전달 매출을 prev_sales 라는 새로운 칼럼에 저장

monthly_sales['growth_rate'] = (monthly_sales['total_sales'] - monthly_sales['prev_sales']) / monthly_sales['prev_sales']

# 4. NaN 제거 후 평균 성장률 계산
avg_growth_by_category = monthly_sales.dropna(subset=['growth_rate']).groupby('category')['growth_rate'].mean()

# monthly_sales.dropna(subset=['growth_rate']):growth_rate 값이 NaN(결측치)인 행 제거, 첫달처럼 이전달이 없어서 성장률을 계산할 수 없는 경우 제외
# .groupby('category') : 남은 데이터에서 카테고리 기준으로 그룹화(화장품, 가구 ,식품으로 묶기)
#['growth_rate'].mean(): 각 카테고리별로 성장률의 평균값 계산 즉, 월별 성장률을 평균내서 카테고리별 성장률 구하기

# 5. 결과 출력
print("카테고리별 평균 월별 성장률:")
print(avg_growth_by_category.sort_values(ascending=False))


# year_month → datetime으로 변환
monthly_sales['year_month_dt'] = monthly_sales['year_month'].dt.to_timestamp()

# 영어 월 이름 추출
monthly_sales['month_label'] = monthly_sales['year_month_dt'].dt.strftime('%b')

# 월 순서 지정
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
monthly_sales['month_label'] = pd.Categorical(monthly_sales['month_label'], categories=month_order, ordered=True)

# 그래프 그리기
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='month_label', y='total_sales', hue='category', marker='o')
plt.title('카테고리별 월별 매출액', fontsize=16)
plt.xlabel('월', fontsize=12)
plt.ylabel('매출액', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 출력 결과를 보고 코딩하세요

# 출력 결과를 보고 코딩하세요
df 
# order_date 기준으로 월단위로 그룹화
df['month'] = df['order_date'].dt.to_period('M')

# 월별 판매량 집계
# 상품별 합계 총합
monthly_sales = df.groupby(['category','product_name','month'])['total_sales'].sum().reset_index() # 인덱스를 일반커럼으로 돌려서 보기 쉽게

# 상위10% 기준값 계산
threshold = monthly_sales['total_sales'].quantile(0.9) # 상위 10% 에 해당하는 판매량 기준값 계산

# 상위10%  이상인 행만 필터링
top_10_percent = monthly_sales[monthly_sales['total_sales'] >= threshold]
print(top_10_percent)

# 각 카테고리별로 전체에서 몇 프로차지하는지
#1.전체항목수를 카테고리별로 계산
total_counts = monthly_sales['category'].value_counts()

#2.상위 10% 항목수 카테고리별로 계산
top_counts = top_10_percent['category'].value_counts()

#3.비율계산
category_percent = (top_counts / total_counts *100).round(2)

#비율계산한거를 원래데이터프레임인 top_10_percent 추가하기
top_10_percent['category_percent'] = top_10_percent['category'].map(category_percent)
print(top_10_percent)


# # 시각화
# pivot = top_10_percent.pivot_table(          #top_10_percent 데이터 프레임에서 피벗테이블 생성
#     index='month', # 행인덱스를 월 즉, 각 행은 하나의 월 나타냄
#     columns='category', # 열은 category 설정 즉, 하나의 열은 하나의 제품 카테고리 나타냄
#     values='category_percent', # 셀의 값 즉, 각 월-카테고리조합에 대해 해당비율 보여줌
#     aggfunc='mean'  # 같은 월에 여러 제품이 있을 수 있으므로 평균 사용
# ).fillna(0) # 비어있는 셀은 0 사용

# 그래프 그리기



pivot = top_10_percent.pivot_table(
    index='month',
    columns='category',
    values='category_percent',
    aggfunc='mean'
).fillna(0)

# 월 기준 정렬
pivot = pivot.sort_index()

# 시각화
plt.figure(figsize=(14, 7))
pivot.plot.area(alpha=0.8, colormap='tab10')

plt.title('월별 상위 10% 상품의 카테고리 분포 변화', fontsize=16)
plt.xlabel('월', fontsize=12)
plt.ylabel('카테고리 비율 (%)', fontsize=12)
plt.legend(title='카테고리', loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()



```

```python
# copilot이 도와줌


import matplotlib.pyplot as plt
import matplotlib as mpl

# 한글 폰트 설정 (예: 맑은 고딕)
mpl.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 깨짐 방지

#1.고객별로 주문날짜 기준으로 정렬하고 구매순서에rank부여
df['order_date'] = pd.to_datetime(df['order_date'])
df = df.sort_values(by=['customer_id','order_date'])
df['rank']=df.groupby('customer_id')['order_date'].rank(method='dense').astype(int)

first = df[df['rank']==1][['customer_id','category']].rename(columns={'category':'first_category'})
second = df[df['rank']==2][['customer_id','category']].rename(columns={'category':'second_category'})

df_transition = pd.merge(first,second)

# 1.전환조합별로 집계 및 비율 계산
transition_counts = df_transition.groupby(['first_category','second_category']).size().reset_index(name='count')

#비율계산
# transition_counts['ratio'] = (transition_counts['count']/transition_counts['count'].sum()).round(3)

# 예)'의류'에서 시작한 고객중 몇 % 가 '의류'로 두번째 구매했는지
# transition_counts['ratio'] = (transition_counts['count']/transition_counts['count'].sum()).round(3)

# 2.첫 구매 카테고리별 총합계산
total_by_first = transition_counts.groupby('first_category')['count'].transform('sum')
print(total_by_first)

#<핵심>
# .groupby('first_category') : 첫구매 카테고리 기준으로 그룹을 나눔
# 즉, '의류'로 시작한 모든 전환, '화장품'으로 시작한 모든 전환

# 3.비율계산: 각 첫구매 카테고리 기준으로 두번째 구매 비율
transition_counts['ratio'] = (transition_counts['count']/total_by_first).round(3)
transition_counts

# 시각화
# 1.피벗테이블로 변환
pivot_table = transition_counts.pivot(index='first_category',columns='second_category',values='ratio')

# 2.히트맵그리기
plt.figure(figsize=(10,8))
sns.heatmap(pivot_table,annot=True,fmt=".3f",cmap="YlGnBu",linewidths=0.5)

# 3.제목 및 축설정
plt.title('첫 구매 → 두 번째 구매 카테고리 전환 비율',fontsize=14)
plt.show()

```
